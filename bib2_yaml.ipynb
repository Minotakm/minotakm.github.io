```python
import re
from datetime import datetime

def parse_author(author):
    parts = author.split()
    if len(parts) == 2:
        return f"{parts[1]} {parts[0]}"
    else:
        return author

def bib_to_yaml(bibtex_entry):
    # Regular expressions to extract information from BibTeX entry
    title = re.search(r'title = \{(.+?)\}', bibtex_entry).group(1)
    authors = [parse_author(match.group(1)) for match in re.finditer(r'author = \{(.+?)\}', bibtex_entry)]
    year = re.search(r'year = \{(\d{4})\}', bibtex_entry).group(1)
    doi = re.search(r'doi = \{(.+?)\}', bibtex_entry).group(1)
    
    # Create YAML content
    yaml_content = "---\n"
    yaml_content += f"title: \"{title}\"\n"
    yaml_content += "authors:\n"
    yaml_content += "\n".join([f"  - {author}" for author in authors]) + "\n"
    yaml_content += "author_notes: []\n"
    yaml_content += f"date: \"{year}\"\n"
    yaml_content += f"doi: \"{doi}\"\n"
    yaml_content += f"publishDate: \"{datetime(int(year), 8, 22, 0, 0, 0).isoformat()}\"\n"
    yaml_content += "publication_types: [\"article-journal\"]\n"
    yaml_content += "publication: \"*Physical Review Materials (7)*\"\n"
    yaml_content += "publication_short: \"\"\n"
    yaml_content += "abstract: 'The prediction of phase diagrams in the search for new phases is a complex and computationally intensive task... (Your abstract here)'\n"
    yaml_content += "summary: ''\n"
    yaml_content += "tags:\n"
    yaml_content += "- Interatomic & molecular potentials\n"
    yaml_content += "- Density Functional Theory\n"
    yaml_content += "- Machine Learning\n"
    yaml_content += "featured: false\n"
    yaml_content += f"url_pdf: 'https://journals.aps.org/prmaterials/abstract/{doi}'\n"
    yaml_content += "url_code: ''\n"
    yaml_content += "url_dataset: ''\n"
    yaml_content += "url_poster: ''\n"
    yaml_content += "url_project: ''\n"
    yaml_content += "url_slides: ''\n"
    yaml_content += "url_source: ''\n"
    yaml_content += "url_video: ''\n"
    yaml_content += "image:\n"
    yaml_content += "  caption: 'Image credit: [**Unsplash**](https://unsplash.com/photos/jdD8gXaTZsc)'\n"
    yaml_content += "  focal_point: \"\"\n"
    yaml_content += "  preview_only: True\n"
    yaml_content += "projects: []\n"
    yaml_content += "---\n"

    return yaml_content



bibtex_entry = """
@article{rossignol2023machine,
  title={Machine-Learning-Assisted Construction of Ternary Convex Hull Diagrams},
  author={Rossignol, Hugo and Minotakis, Michail and Cobelli, Matteo and Sanvito, Stefano},
  journal={arXiv preprint arXiv:2308.15907},
  year={2023}
}
"""

title = re.search(r'title = \{(.+?)\}', bibtex_entry).group(1)
year = re.search(r'year = \{(\d{4})\}', bibtex_entry).group(1)



yaml_content = bib_to_yaml(bibtex_entry)


with open(f"{'_'.join(title.split()) +'_' +year}.md", 'w') as fout:
    for i in yaml_content:
        fout.write(f'{i}')
```
